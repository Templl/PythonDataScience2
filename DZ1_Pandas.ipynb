### Задание 4
Создайте датафрейм authors_stat на основе информации из authors_price. В датафрейме authors_stat должны быть четыре столбца:
author_name, min_price, max_price и mean_price,
в которых должны содержаться соответственно имя автора,минимальная, максимальная и средняя цена на книги этого автора.


```python
authors_price
```

```python
groupby = authors_price.groupby("author_name")
max_price = authors_price.groupby("author_name").agg({'price':'max'})
#authors_stat = groupby.agg({"price": "max", "total": "count"})
#authors_stat 
max_price = max_price.rename(columns={'price':'max_price'})
max_price
```

```python
min_price = groupby.agg({"price": "min"})
min_price = min_price.rename(columns={'price':'min_price'})
min_price
```

```python
mean = groupby.agg({"price": "mean"})
mean = mean.rename(columns={'price':'mean'})
mean
```

```python
mean
```

```python
authors_stat = pd.merge(max_price,min_price, on='author_name', how='left')
authors_stat
```

```python
authors_stat = pd.merge(authors_stat,mean, on='author_name', how='outer')
```
