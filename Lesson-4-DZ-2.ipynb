### Задание 2


```python
#С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.
t = np.linspace(0, 10, 51)
```

```python
#Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.
f = np.cos(t)
```

```python
#Постройте линейную диаграмму, используя массив t для координат по горизонтали, 
#а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.

label_font = {
    "fontsize": 9,
    "family": "serif",
}
plt.plot(t,f,color="green")
#Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 
#'Значения t' и для вертикальной - 'Значения f'.
plt.title("График f(t)",fontweight="bold",)
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f",fontdict=label_font) 
#Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.
plt.axis([0.5,9.5,-2.5,2.5])
plt.show()
```

### *Задание 3

```python
#С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
x = np.linspace(-3, 3, 51)
x
```

```python
#Создайте массивы y1, y2, y3, y4 по следующим формулам:
#y1 = x**2
#y2 = 2 * x + 0.5
#y3 = -3 * x - 1.5
#y4 = sin(x)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
y4
```

```python
#Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure 
#с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке
#plt.plot(x,y1)
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title("График 1")

ax2.plot(x, y2)
ax2.set_title("График 2")

ax3.plot(x, y3)
ax3.set_title("График 3")

ax4.plot(x, y4)
ax4.set_title("График 4")

plt.show()
```
